```python
import os
import sys
nb_dir = os.path.split(os.getcwd())[0]
if nb_dir not in sys.path:
    sys.path.append(nb_dir)

import pandas as pd
import util.utilities as util
from pybtex.database import parse_file
```

```python
file_to_scan = '../../bib/gsatd.bib'
original='my_atd_extracted.md'
sorted_by_year='my_atd_extracted_sorted_by_year.md'
sorted_by_title='my_atd_extracted_sorted_by_title.md'
file_xls_to_scan = '../../xls/google_scholar_atd.xlsx'

my_data = dict()
my_data['key'] = list()
my_data['year'] = list()
my_data['list_authors'] = list()
my_data['title'] = list()
my_data['abstract'] = list()
my_data['link'] = list()

my_xls_dict = {}

bib_data = parse_file(file_to_scan)
my_xls_data = pd.read_excel(file_xls_to_scan)

for index, row in my_xls_data.iterrows():
    list_content = []
    list_content.append(row['Abstrract'])
    list_content.append(row['Link'])
    my_xls_dict[row['key']] = list_content
```

```python
try:
    i = 1
    for key, value in bib_data.entries.items(): 
        if ('year' in value.fields):
            year = value.fields['year']
        else:
            year = None
        if (value.persons is not None):
            list_authors = util.show_names(value.persons['author'])
        else: 
            list_authors = None
        title = value.fields['title']
        if ('abstract' in value.fields): 
            abstract = value.fields['abstract']
        else: 
            if key in my_xls_dict.keys():
                abstract = my_xls_dict[key][0]
            else: 
                abstract = None
        if my_xls_dict[key][1] is not None: 
            link = my_xls_dict[key][1]
        else: 
            link = None
        my_data['key'].append(key)
        my_data['year'].append(year)
        my_data['list_authors'].append(list_authors)
        my_data['title'].append(title)
        my_data['abstract'].append(abstract)
        my_data['link'].append(link)

        print('{}, {}, {}, {}, {}, {}, {}'.format(i, key, year, list_authors, title, abstract, link))
        print('---')
        i += 1
    print("Sucesso! Leitura de {} entradas no arquivo {}".format(i-1, file_to_scan))
except Exception as e:
    print("Erro " + str(e)+ " na leitura do arquivo " + file_to_scan)
```

```python
df_bib = pd.DataFrame(data=my_data, columns=['key', 'year', 'list_authors', 'title', 'abstract', 'link'])
df_bib[['key', 'year', 'list_authors', 'title', 'abstract', 'link']]
```

```python
util.create_markdown_table(my_df=df_bib, my_path='../../md', my_file_name=original)

df_bib_sort_by_year = df_bib.sort_values('year')
util.create_markdown_table(my_df=df_bib_sort_by_year, my_path='../../md', my_file_name=sorted_by_year)
df_bib_sort_by_year
```

```python
df_bib_sort_by_title = df_bib.sort_values('title')
util.create_markdown_table(my_df=df_bib_sort_by_title, my_path='../../md', my_file_name=sorted_by_title)
df_bib_sort_by_title
```
